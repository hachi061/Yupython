### 將一個DataFrame做成一個統計資料

```python
df1 = df[["年底人口數","面積","人口密度"]]

def statistic(s):
    print(s[s==s.max()].index[0])
    return pd.Series([s.mean(),s.max(),s.min(),s.std(),s.median()],index=['平均','最高','最低','標準差','中間數'])

df1.apply(statistic,axis=0)
```
